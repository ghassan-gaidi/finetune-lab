# 🆚 VS Code Remote Tunnel — Colab

Set up a **VS Code Remote Tunnel** from Colab. Connect from VS Code Desktop or `vscode.dev`.

**Open in Colab:**
[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/ghassan-gaidi/finetune-lab/blob/main/notebooks/setup/colab-server-vscode.ipynb)

---
**How it works:**
1. Downloads VS Code CLI on Colab
2. Authenticates with your GitHub account
3. Creates a named tunnel (`colab-gpu`)
4. Connect from your local VS Code → Remote Explorer → Connect to Tunnel

**⚠️ Requires:** A GitHub or Microsoft account for authentication.

## 1. Download VS Code CLI

````
# Create directories for persistent data
!mkdir -p /content/.vscode-server/cli \
         /content/.vscode-server/data \
         /content/.vscode-server/extensions

# Download VS Code CLI (latest stable for Linux x64)
DOWNLOAD_URL = "https://code.visualstudio.com/sha/download?build=stable&os=cli-alpine-x64"
!wget -q "$DOWNLOAD_URL" -O /content/.vscode-server/vscode-cli.tar.gz
!tar -xzf /content/.vscode-server/vscode-cli.tar.gz -C /content/.vscode-server/cli --strip-components=1
!chmod +x /content/.vscode-server/cli/code

# Verify
!./vscode-server/cli/code --version
````

## 2. Login to GitHub

Run the cell below. It will print a **device code URL** like:
```
https://github.com/login/device
Code: ABCD-1234
```

**Open that URL, enter the code, and authorize.** This links the Colab VS Code tunnel to your GitHub account.

After authorizing, come back here — the cell will continue automatically.

````
# Login (interactive — follow the device code prompt)
!./vscode-server/cli/code tunnel user login --provider github
````

## 3. Start the Tunnel

Once authenticated, this starts the tunnel named **`colab-gpu`**. It runs until you stop the cell.

Connect from your local VS Code:
1. Open VS Code → Activity Bar → **Remote Explorer** (the monitor icon)
2. Under **Remote Tunnels**, click **Connect to Tunnel...**
3. Authenticate with the **same GitHub account**
4. Select the tunnel named `colab-gpu`
5. You're in! Full terminal, file explorer, and GPU access.

````
NAME = "colab-gpu"  # change to something unique if you share accounts

!./vscode-server/cli/code tunnel \
    --accept-server-license-terms \
    --name "$NAME" \
    --server-data-dir /content/.vscode-server/data \
    --extensions-dir /content/.vscode-server/extensions \
    --cli-data-dir /content/.vscode-server/cli-data

# This blocks until you Ctrl+C / stop the cell
````

## 💡 Tips

- **Extensions** persist if you set `--extensions-dir` as above
- **No need to re-login** if `/content/.vscode-server/cli-data` persists
- On a **fresh Colab runtime**, re-run cells 1 and 3 (login persists if you used the same directories)
- Access from **any browser**: `https://vscode.dev/tunnel/colab-gpu/`
- **Pre-install deps** in the terminal after connecting:
  ```
  pip install torch transformers accelerate peft bitsandbytes datasets trl
  ```
